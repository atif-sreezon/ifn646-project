```python
import re

fasta = open('zincfinger_c2h2.fa', 'r').read()

reads = fasta.split('>')[1:]
headers = [read.split('\n')[0] for read in reads]
IDs = [head.split('|')[1] for head in headers]
gene_names = [head.split('|')[2] for head in headers]
seqs = [''.join(read.split('\n')[1:]) for read in reads]

seqs_list = []

for idx, gene_id in enumerate(IDs):
    seq_dict = {}
    seq_dict['Prot_ID'] = gene_id
    seq_dict['Gene_name'] = gene_names[idx]
    seq_dict['Seq'] = seqs[idx]
    
    seqs_list.append(seq_dict)
    
    
query = r'[C][A-Z]{2,4}[C][A-Z]{3}[LIVMFYWC][A-Z]{8}[H][A-Z]{3,5}[H]'

match = [seq for seq in seqs_list if re.search(query, seq['Seq']) != None]

for m in match:
    print('Swiss-Prot ID: {}'.format(m['Prot_ID']))
    print('Gene Name: {}'.format(m['Gene_name']))
    print('Start: {}'.format())
    print('Sequence: {}\n'.format(m['Seq']))
```
